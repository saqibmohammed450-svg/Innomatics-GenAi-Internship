*_1: Employee Performance Bonus Eligibility_*

```python
# Employee performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest performance score
highest_score = max(employees.values())

# Step 2: Find all employees with the highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 3: Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), "(Score:", highest_score, ")")
```

*_2: Search Query Keyword Analysis_*

```python
query = "Buy mobile phone buy phone online"

# Normalize text
query = query.lower()

# Tokenize
words = query.split()

# Count keyword frequency
keyword_frequency = {}

for word in words:
    keyword_frequency[word] = keyword_frequency.get(word, 0) + 1

# Filter keywords searched more than once
repeated_keywords = {
    word: count
    for word, count in keyword_frequency.items()
    if count > 1
}

print(repeated_keywords)
```

*_3: Sensor Data Validation_*

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

*_ 4: Email Domain Usage Analysis_*

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count users per domain
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_users = len(emails)

# Calculate and display percentage usage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

*_5: Sales Spike Detection_*

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average daily sales
average_sales = sum(sales) / len(sales)

# Detect sales spikes (more than 30% above average)
threshold = average_sales * 1.3

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

*_ 6: Duplicate User ID Detection_*

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Count occurrences of each user ID
for user_id in user_ids:
    id_count[user_id] = id_count.get(user_id, 0) + 1

# Display only duplicate IDs
for user_id, count in id_count.items():
    if count > 1:
        print(f"{user_id} → {count} times")
```

*_Problem 1: Unique Words in a Sentence_*

```python
# Taking sentence input from user
sentence = input("Enter sentence: ")

# Step 1: Split sentence into words
words = sentence.split()

# Step 2: Convert list to set to get unique words
unique_words = set(words)

# Step 3: Display result
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

*_Problem 2: Highest Salary from Employee Data_*

```python
# Employee data stored in a dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_salary = 0
highest_paid_employee = ""

for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_paid_employee = name

# Display result
print("Highest Salary:", highest_paid_employee, "-", highest_salary)
```

*_Problem 3: Find Maximum and Minimum Values_*

```python
# Given list of numbers
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Display results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

*_Problem 4: Count Products Above a Price Threshold_*

```python
# List of product prices
prices = [450, 1200, 899, 1500, 300]

count = 0

# Count products priced above 1000
for price in prices:
    if price > 1000:
        count += 1

# Display result
print("Products above 1000:", count)
```

*_Problem 5: Calculate Attendance Percentage_*

```python
# Attendance list
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

# Count number of present days
for day in attendance:
    if day == "P":
        present_count += 1

# Calculate attendance percentage
total_days = len(attendance)
attendance_percentage = (present_count / total_days) * 100

# Display result
print("Attendance Percentage:", attendance_percentage)
```

*_Problem 7: Count Character Frequency_*

```python
# Given string
text = "python"

# Dictionary to store character frequency
char_frequency = {}

# Count each character
for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1

# Display result
print(char_frequency)
```

*_Problem 10: Calculate Average Marks_*

```python
# List of marks
marks = [70, 75, 80, 65]

total_marks = 0

# Calculate total marks
for m in marks:
    total_marks += m

# Calculate average
average_marks = total_marks / len(marks)

# Display result
print("Average Marks:", average_marks)
```

**_1: Smart Light Controller_**

```python
# Class to control a smart light
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"
        print(f"{self.name} is {self.status}")

    def turn_off(self):
        self.status = "OFF"
        print(f"{self.name} is {self.status}")


# Sample usage
light_name = "Bedroom Light"
action = "ON"

light = SmartLight(light_name)

if action == "ON":
    light.turn_on()
else:
    light.turn_off()
```

**_ 2: Employee ID Card System_**

```python
# Class to represent an Employee ID Card
class EmployeeIDCard:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample usage
employee_name = "Rahul"
employee_id = "EMP102"
department = "IT"

employee = EmployeeIDCard(employee_name, employee_id, department)
employee.display_id_card()
```

**_ 3: Mobile Contact Record_**

```python
# Class to store mobile contact details
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample usage
contact_name = "Anita"
phone_number = 9876543210

contact = Contact(contact_name, phone_number)
contact.display_contact()
```

**_ 4: Product Price Tag Generator_**

```python
# Class to generate product price tag
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Sample usage
product_name = "Headphones"
product_price = 2499

product = Product(product_name, product_price)
product.display_price_tag()
```

**_5: Movie Rating Display System_**

```python
# Class to store and display movie rating details
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie_details(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Sample usage
movie_name = "Inception"
movie_rating = 4.8

movie = Movie(movie_name, movie_rating)
movie.display_movie_details()
```

**_6: Delivery Address Manager_**

```python
# Class to manage delivery address details
class DeliveryAddress:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Sample usage
customer_name = "Suman"
delivery_address = "Hyderabad"

delivery = DeliveryAddress(customer_name, delivery_address)
delivery.display_delivery_details()
```

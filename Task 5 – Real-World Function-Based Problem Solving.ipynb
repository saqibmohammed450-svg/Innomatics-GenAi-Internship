**_ 1: Social Media – Post Engagement Analyzer_**

```python
# Function to analyze post engagement
def analyze_engagement(likes):
    total_likes = 0

    # Calculate total likes using loop
    for like in likes:
        total_likes += like

    # Determine post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Example likes data
post_likes = [200, 350, 150, 300, 50]

# Analyze engagement
total, post_status = analyze_engagement(post_likes)

# Display result
print("Total Likes:", total)
print("Post Status:", post_status)
```

**__3: Agriculture – Rainfall Adequacy Checker__**



```python
# Function to analyze rainfall data
def check_rainfall(rainfall_data, required_level):
    total_rainfall = 0

    # Calculate total rainfall using loop
    for value in rainfall_data:
        total_rainfall += value

    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)

    # Determine rainfall status
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average_rainfall, status


# Example rainfall data (in mm)
seasonal_rainfall = [70, 75, 68, 74, 73]
required_rainfall_level = 70

# Analyze rainfall
average, rainfall_status = check_rainfall(seasonal_rainfall, required_rainfall_level)

# Display result
print("Average Rainfall:", int(average))
print("Rainfall Status:", rainfall_status)
```

**__4: Social Media – Duplicate Account Detection__**

```python
# Function to detect duplicate usernames
def detect_duplicate_accounts(usernames):
    if len(usernames) != len(set(usernames)):
        return "Yes"
    else:
        return "No"


# Example username data
usernames = ["alice", "bob", "alice", "charlie"]

# Check for duplicates
result = detect_duplicate_accounts(usernames)

# Display result
print("Duplicate Accounts Found:", result)
```

**_6: Agriculture – Premium Crop Price Filter_**

```python
# Function to filter premium crop prices
def filter_premium_crops(prices):
    premium_crops = []

    # Filter prices using loop
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    return premium_crops


# Example crop prices
crop_prices = [1800, 2500, 3200, 1500]

# Get premium crops
premium = filter_premium_crops(crop_prices)

# Display result
print("Premium Crops:", premium)
```

**__7: System Monitoring – Application Health Checker__**

**_9: E-Learning – Student Attendance Eligibility System_**

```python
# Function to calculate attendance percentage and check eligibility
def check_exam_eligibility(attendance):
    present_days = 0

    # Count present days using loop
    for day in attendance:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    attendance_percentage = (present_days / len(attendance)) * 100

    # Determine eligibility
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return attendance_percentage, status


# Example attendance data
student_attendance = ["P", "P", "A", "P", "P"]

# Check eligibility
percentage, eligibility = check_exam_eligibility(student_attendance)

# Display result
print("Attendance Percentage:", percentage)
print("Exam Eligibility:", eligibility)
```

**_10: Smart Electricity Bill Analyzer_**

```python
# Function to calculate electricity bill and usage classification
def calculate_electricity_bill(units):
    bill = 0

    # Billing slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Example usage
units_consumed = 250

total_bill, usage_status = calculate_electricity_bill(units_consumed)

# Display result
print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Status:", usage_status)
```

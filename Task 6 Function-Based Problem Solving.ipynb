*_1:Smart Parking Lot Management System_*

```python
# Function to manage smart parking lot
def manage_parking(vehicle_logs, capacity):
    parked_vehicles = 0
    peak_usage = 0

    # Process vehicle entry and exit logs
    for log in vehicle_logs:
        if log == "IN":
            parked_vehicles += 1
        elif log == "OUT" and parked_vehicles > 0:
            parked_vehicles -= 1

        # Track peak parking usage
        if parked_vehicles > peak_usage:
            peak_usage = parked_vehicles

    # Determine parking status
    if parked_vehicles > capacity:
        status = "Alert: Parking Exceeds Capacity"
    else:
        status = "Available"

    return parked_vehicles, peak_usage, status


# Sample input
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# Manage parking
current, peak, parking_status = manage_parking(vehicle_logs, parking_capacity)

# Display result
print("Currently Parked Vehicles:", current)
print("Parking Status:", parking_status)
```

*_2:Online Food Delivery Time Estimator_*

```python
# Function to estimate delivery time
def estimate_delivery_time(distance_km, traffic_level, weather):
    # Base speed assumption: 1 km = 5 minutes
    time = distance_km * 5

    # Traffic-based delay
    if traffic_level == "High":
        time += 15
    elif traffic_level == "Medium":
        time += 8

    # Weather-based delay
    if weather == "Rainy":
        time += 10
    elif weather == "Stormy":
        time += 20

    return time


# Sample input
distance = 8
traffic = "High"
weather_condition = "Rainy"

# Calculate ETA
eta = estimate_delivery_time(distance, traffic, weather_condition)

# Display result
print("Estimated Delivery Time:", eta, "minutes")
```

*_3:Movie Theatre Seat Occupancy Analyzer_*

```python
# Function to analyze seat occupancy
def analyze_seat_occupancy(total_seats, booked_seats):
    booked_count = 0

    # Count booked seats using loop
    for seat in booked_seats:
        if seat == 1:
            booked_count += 1

    # Calculate occupancy percentage
    occupancy_percentage = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy_percentage == 100:
        status = "Housefull"
    elif occupancy_percentage >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"

    return occupancy_percentage, status


# Sample input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

# Analyze occupancy
occupancy, show_status = analyze_seat_occupancy(total_seats, booked_seats)

# Display result
print("Occupancy:", int(occupancy), "%")
print("Show Status:", show_status)
```

*_ 4:Cloud Server Load Classification System_*

```python
# Function to classify server load based on CPU usage
def classify_server_load(cpu_readings):
    total = 0

    # Calculate total CPU usage using loop
    for usage in cpu_readings:
        total += usage

    # Calculate average CPU load
    average_cpu = total / len(cpu_readings)

    # Classify server load
    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    return average_cpu, status


# Sample input
cpu_readings = [45, 60, 70, 85, 90]

# Classify server load
average, server_status = classify_server_load(cpu_readings)

# Display result
print("Average CPU Load:", int(average), "%")
print("Server Status:", server_status)
```

*_5:Smart Classroom Resource Usage Monitor_*

```python
# Function to monitor classroom resource usage
def monitor_resource_usage(resource_usage):
    overused_resources = []
    usage_limit = 8  # hours threshold

    # Check each resource usage
    for resource, hours in resource_usage.items():
        if hours > usage_limit:
            overused_resources.append(resource)

    # Determine energy alert status
    if overused_resources:
        energy_alert = "Yes"
    else:
        energy_alert = "No"

    return overused_resources, energy_alert


# Sample input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Monitor resource usage
overused, alert = monitor_resource_usage(resource_usage)

# Display result
print("Overused Resources:", ", ".join(overused))
print("Energy Alert:", alert)
```

*_6: Online Event Registration Capacity Controller_*

```python
# Function to manage event registrations
def manage_event_registrations(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # Process each registration using loop
    for _ in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Determine registration status
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"

    return confirmed, waitlisted, status


# Sample input
event_capacity = 100
total_registrations = 105

# Manage registrations
confirmed_count, waitlisted_count, registration_status = manage_event_registrations(
    event_capacity, total_registrations
)

# Display result
print("Confirmed Registrations:", confirmed_count)
print("Waitlisted Users:", waitlisted_count)
print("Registration Status:", registration_status)
```
